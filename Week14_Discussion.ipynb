## Reverse Egineering

According to Amazon: 

"Unlike traditional collaborative
filtering, our algorithm’s online computation scales
independently of the number of customers and
number of items in the product catalog. Our algorithm  produces  recommendations  in  realtime,
scales to massive data sets, and generates high 
quality recommendations."(https://datajobs.com/data-science-repo/Recommender-Systems-[Amazon].pdf)

This differ somewhat from the traditional collaborative methods: collaborative filtering, clustering, search-based methods. 

Also, customer can go alter their recommendations on the Amazon website. Amazon uses email marketing to send user recommendations where the user can click on the link from their email and when in the recommendations section, can alter and rate their recommendations and recommendations settings.
